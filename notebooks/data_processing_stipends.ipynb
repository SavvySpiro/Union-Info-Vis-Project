```python
pathname = "data/cleaned_stipends.csv"
stipends = pd.read_csv(pathname)
```

```python
stipends["University"].unique()
```

```python
# Subselection of Boston-area universities that we're using for comparison
universities = ["Harvard University (HU)", "Northeastern University (NU)", 
                "Northeastern University", 
                "Massachusetts Institute of Technology (MIT)", "MIT", 
                "Tufts University (TU)", "Boston University (BU)", 
                "Boston University ", "Boston University",
                "University of Massachusetts - Boston (UMB)"]
```

```python
boston = stipends[stipends["University"].isin(universities)].copy().reset_index()
```

```python
boston = boston.drop(columns=['index'])
```

```python
def consolidate(uni):
    # Since some of the unis had multiple forms of their name, 
    # I wanted to make them consistent
    if uni == "Northeastern University (NU)":
        return "Northeastern University"
    if uni == "Harvard University (HU)":
        return "Harvard University"
    if uni == "Massachusetts Institute of Technology (MIT)":
        return "MIT"
    if uni == "Boston University (BU)" or uni == "Boston University ":
        return "Boston University"
    if uni == "Tufts University (TU)":
        return "Tufts University"
    if uni == "University of Massachusetts - Boston (UMB)":
        return "UMass Boston"
    return uni

boston["University"] = boston["University"].apply(consolidate)
```
