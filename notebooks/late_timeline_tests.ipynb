```python
import pandas as pd
import plotly.express as px

import plotly.graph_objects as go
import numpy as np
```

```python
test = [["Fuck", 3], ["Fuck", 3], ["Fuck", 3], ["Fuck", 5], ["This", 2], ["This", 7], ["This", 2], ["This", 2]]
temp = pd.DataFrame(test, columns=["Text", "Number"])
```

```python
change_values = temp[["Text", "Number"]].groupby("Text").max().to_dict()['Number']
```

```python
change_values
```

```python
temp['Opacity'] = temp.apply(
    lambda x: x['Number']/change_values[x['Text']], axis=1
)
```

```python
temp
```

```python
temp['Opacity'].to_list()
```
